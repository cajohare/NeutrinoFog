```python
import sys
sys.path.append('../src')
from numpy import *
import matplotlib.pyplot as plt
from PlotFuncs import *
from Like import *
import cmasher as cmr
from matplotlib.colors import LinearSegmentedColormap

pek = line_background(6,'k')
interval = flipud(linspace(0.0,1.0))
colors = cm.terrain(interval)
cmap = LinearSegmentedColormap.from_list('name', colors)

vmax = 7
vmin = 2.2
```

```python
fig,ax = MakeLimitPlot_SI(Collected=True,alph=1,edgecolor=col_alpha('gray',0.75),facecolor=col_alpha('gray',0.5),ymin=1e-50,ymax=1e-41)

data = loadtxt('../data/WIMPLimits/mylimits/DLNuFloorXe_detailed_SI.txt')
m,sig,NUFLOOR,DY = Floor_2D(data,filt=True,filt_width=2,Ex_crit=1e10)
cnt = plt.contourf(m,sig,DY,levels=linspace(2,15,100),vmax=vmax,vmin=vmin,cmap=cmap)
for c in cnt.collections:
    c.set_edgecolor("face")
plt.plot(m,NUFLOOR,'-',color='brown',lw=3,path_effects=pek,zorder=100)

plt.gcf().text(0.82,0.9,r'$\left(\frac{{\rm d}\ln\sigma}{{\rm d}\ln N}\right)^{-1}$',fontsize=35,color='w')
plt.gcf().text(0.15*(1-0.01),0.16*(1+0.01),r'{\bf Xenon}',color='k',fontsize=50,alpha=0.2)
plt.gcf().text(0.15,0.16,r'{\bf Xenon}',color='brown',fontsize=50)

MySaveFig(fig,'NuFloorDetailed_Xe_SI')
```
